```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
from datetime import datetime
import pytz
from timezonefinder import TimezoneFinder
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Send a GET request to the Wikipedia URL
url = "https://en.wikipedia.org/wiki/2024_Formula_One_World_Championship"
response = requests.get(url)

# Parse the HTML content
soup = BeautifulSoup(response.text, 'html.parser')

# Find the section containing the desired sub-theme
section_heading = soup.find('span', {'id': 'World_Drivers.27_Championship_standings'})
section = section_heading.find_parent()
```

```python
# Initialize lists to store data
data = []

# Extract all table rows under the sub-theme "World Drivers' Championship standings"
table = section.find_next('table', class_='wikitable')
if table:
    rows = table.find_all('tr')
    for row in rows:
        cells = row.find_all(['th', 'td'])
        if cells:
            data.append([cell.get_text(strip=True) for cell in cells])

# If the table header is not extracted properly, try to get the header from the first row
if not data:
    header_row = table.find('tr')
    header_cells = header_row.find_all(['th', 'td'])
    header = [cell.get_text(strip=True) for cell in header_cells]
    data.append(header)

# Convert the data into a pandas DataFrame
drivers_standing = pd.DataFrame(data[1:], columns=data[0])

# Remove the last two rows
drivers_standing = drivers_standing.drop([21, 22])

# Selecting the first, second, and last columns
drivers_standing = drivers_standing.iloc[:, [0, 1, -1]]

# Convert the "Points" column to integer
drivers_standing["Points"] = drivers_standing["Points"].astype(int)

# Display the DataFrame
print("Current Drivers' Standings")
display(drivers_standing)
```

```python
# Parse the HTML content
soup = BeautifulSoup(response.text, 'html.parser')

# Find the section containing the desired sub-theme "Calendar"
section_heading = soup.find('span', {'id': 'Calendar'})
section = section_heading.find_parent()

# Initialize lists to store data
data = []

# Extract all table rows under the "Calendar" section
table = section.find_next('table', class_='wikitable')
if table:
    rows = table.find_all('tr')
    for row in rows:
        cells = row.find_all(['th', 'td'])
        if cells:
            data.append([cell.get_text(strip=True) for cell in cells])

# If the table header is not extracted properly, try to get the header from the first row
if not data:
    header_row = table.find('tr')
    header_cells = header_row.find_all(['th', 'td'])
    header = [cell.get_text(strip=True) for cell in header_cells]
    data.append(header)

# Convert the data into a pandas DataFrame
calendar = pd.DataFrame(data[1:], columns=data[0])

# Remove the last row
calendar = calendar.drop([24])

# Convert the race date to yyyy-mm-dd format
calendar['Race date'] = calendar['Race date'].apply(lambda x: datetime.strptime(x, '%d %B').replace(year=datetime.now().year).strftime('%Y-%m-%d'))

# Add a column titled "Type"
calendar['Type'] = calendar['Grand Prix'].apply(lambda x: 'Sprint' if any(keyword in x for keyword in ['China', 'Miami', 'Austria', 'United States', 'Brazil', 'Qatar']) else 'Race')

# Display the DataFrame
print("Calendar of Races")
display(calendar)
```

```python
# Function to find all upcoming races
def find_upcoming_races(calendar):
    current_date = pd.to_datetime(datetime.now().date())  # Current date as datetime64[ns]

    # Convert the 'Race date' column to datetime
    calendar['Race date'] = pd.to_datetime(calendar['Race date'])

    # Find all upcoming races
    upcoming_races = calendar[calendar['Race date'] > current_date].sort_values(by='Race date')

    if not upcoming_races.empty:
        print("Upcoming races details:")
        display(upcoming_races)
        return upcoming_races
    else:
        print("No upcoming races.")
        return None

# Find all upcoming races
upcoming_races = find_upcoming_races(calendar)
```

```python
# Points_distribution
points_distribution = {
    'Race': [25, 18, 15, 12, 10, 8, 6, 4, 2, 1, 0],
    'Sprint': [8, 7, 6, 5, 4, 3, 2, 1, 0]
}

# Function to simulate the remaining races and estimate the probability
def race_simulation(drivers_standing, calendar, num_simulations=10000):
    def single_simulation():
        first_place_points = drivers_standing['Points'].iloc[0]
        second_place_points = drivers_standing['Points'].iloc[1]
        second_place_pass = 0

        for _ in range(num_simulations):
            first_points = first_place_points
            second_points = second_place_points

            for _, race in calendar[calendar['Race date'] > datetime.now()].iterrows():
                race_type = race['Type']

                if race_type == 'Race':
                    first_points += np.random.choice(points_distribution['Race'])
                    second_points += np.random.choice(points_distribution['Race'])
                elif race_type == 'Sprint':
                    # Add points from Sprint and Race points to both drivers
                    first_points += np.random.choice(points_distribution['Race']) + np.random.choice(points_distribution['Sprint'])
                    second_points += np.random.choice(points_distribution['Race']) + np.random.choice(points_distribution['Sprint'])

                # Randomly assign the fastest lap point
                fastest_lap_award = np.random.choice([0, 1, 2], p=[0.5, 0.25, 0.25])  # 50% chance to neither, 25% chance to first or second
                if fastest_lap_award == 1:
                    first_points += 1
                elif fastest_lap_award == 2:
                    second_points += 1

            if second_points > first_points:
                second_place_pass += 1

        return second_place_pass / num_simulations

    # Repeat the simulation 10 times and compute the average probability
    probabilities = [single_simulation() for _ in range(10)]
    average_probability = np.mean(probabilities)

    return average_probability

# Run the simulation using all upcoming races
if upcoming_races is not None:
    num_simulations = 1000  # Number of simulations
    probability = race_simulation(drivers_standing, upcoming_races, num_simulations)

    # Extract first and second place driver names
    first_place_driver = drivers_standing['Driver'].iloc[0]
    second_place_driver = drivers_standing['Driver'].iloc[1]

    print(f"Probability that {second_place_driver} overtakes {first_place_driver}: {probability:.2%}")
else:
    print("No upcoming races to simulate.")
```
